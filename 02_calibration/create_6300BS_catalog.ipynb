```python
import numpy as np
import time
```

```python
from astroquery.simbad import Simbad
# Simbad.add_votable_fields('sptype','flux(U)','flux(B)','flux(V)','flux(R)','flux(I)','flux(J)','flux(H)','flux(K)')
Simbad.add_votable_fields('sptype','flux(V)')
from astropy import units as u
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
```

```python
thedata = np.genfromtxt(
    'Haritonov_catalog_6300_filt.csv',           # file name
    skip_header=1,          # lines to skip at the top
    skip_footer=0,          # lines to skip at the bottom
    delimiter=',',          # column delimiter
    dtype='int32',        # data type
    filling_values=0,)       # fill missing values with 0
    
```

```python
print(np.size(thedata,0))
print(np.size(thedata,1))
```

```python
thedata[0,:]
```

```python
BS_flux=np.zeros(np.size(thedata,0))
for i in range(np.size(thedata,0)):
    BS_flux[i]=(float(thedata[i,2])*(10**-float(thedata[i,3]))+float(thedata[i,4])*(10**-float(thedata[i,5])))/2.0
```

```python
BS_names=[]
for i in range(np.size(thedata,0)):
    BS_names.append('BS ' + str(thedata[i,1]))
```

```python
BS_magv=np.zeros(np.size(thedata,0))
BS_ra=np.zeros(np.size(thedata,0))
BS_dec=np.zeros(np.size(thedata,0))
BS_sptype=[]
for i in range(np.size(thedata,0)):
    print(i)
    if BS_names[i]!='BS 7592':
        sinbad_table=Simbad.query_object(BS_names[i])
    else:
        sinbad_table=Simbad.query_object('13 Vul A')
    
    BS_magv[i]=sinbad_table[0][14]
    BS_sptype.append(str(sinbad_table[0][11])[2:4])
    c=SkyCoord(sinbad_table[0][1] + ' ' + sinbad_table[0][2], unit=(u.hourangle, u.deg), frame='icrs')
    BS_ra[i]=c.ra.deg
    BS_dec[i]=c.dec.deg
```

```python
CAT_FIELDS=['HARITONOV_ID','BS_ID','RA (icrs, deg)','DEC (icrs, deg)','MAG_V','6300 Flux (erg/(sm^2 * s *  sm))']
CAT_6300BS=np.zeros((np.size(thedata,0),6))
for i in range(np.size(thedata,0)):
    CAT_6300BS[i,0]=i+1
    CAT_6300BS[i,1]=thedata[i,1]
    CAT_6300BS[i,2]=BS_ra[i]
    CAT_6300BS[i,3]=BS_dec[i]
    CAT_6300BS[i,4]=BS_magv[i]
    CAT_6300BS[i,5]=BS_flux[i]
```

```python
CAT_6300BS[0,:]
```

```python
file_to_save='Haritonov_6300BS_catalog.npz'
np.savez(file_to_save, CAT_6300BS= CAT_6300BS, CAT_FIELDS=CAT_FIELDS, CAT_SPTYPE=BS_sptype)
```
